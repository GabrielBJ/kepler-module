```python
import odekepler.kepler as kp
```

```python
# initialise the system 
initial_system = kp.initialise_system(0.01671, 1, './outputs/init_system.png')
```

```python
# integrate the system 
integrated_orbit = kp.RKIntegrate.RK4(initial_system, 0.01, "./outputs/integrated_orbit.txt")
```

```python
# generating the animation 

animation = kp.animation_orbit(integrated_orbit, 0.01671, "./outputs/animation.gif")
```
